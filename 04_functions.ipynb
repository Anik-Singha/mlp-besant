#### Functions
- A functions is a block of reusable code that performs a specific task.
- Functions make your program more organized, readble and reduce repetition.

#### Types of functiions-
1. Built-in functions
    - Already available in python (eg: print(),len() )
2. User defined functions:
    - functions created by the user using the def keyword
3. Lambda function
    - anonymous(nameless) functions written in a single line using the lambda keyword

#### User-defined function
- Defining the function
Syntax -
def function_name(p1,p2...):
    block of code
    return value
- Call the function
function_name(p1,p2....)

### Different Ways of declaring User-defined function
- function without input and without return value
- func with input and without return value
- func without input and with return value
- func with input and with return value

```python
# WAP to perform simple calculater +,-,*,/ on 2 numbers
def calculation(a,b):
    x = a
    y = b
    inp = input("Choose 1. Addition, 2. Subtraction, 3. Multiplication, 4. Division")
    if inp == '1':
        print(f"Addition: {a} + {b} = {a+b}")
    elif inp == '2':
        print(f"Subtraction: {a} - {b} = {a-b}")
    elif inp == '3':
        print(f"Multiplication: {a} * {b} = {a*b}")
    elif inp == '4':
        print(f"Division: {a} / {b} = {a/b}")
    else:
        print("Invalid input")
```

```python
calculation(10,5)
```
