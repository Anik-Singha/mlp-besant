#### Functions
- A functions is a block of reusable code that performs a specific task.
- Functions make your program more organized, readble and reduce repetition.

#### Types of functiions-
1. Built-in functions
    - Already available in python (eg: print(),len() )
2. User defined functions:
    - functions created by the user using the def keyword
3. Lambda function
    - anonymous(nameless) functions written in a single line using the lambda keyword

#### User-defined function
- Defining the function
Syntax -
def function_name(p1,p2...):
    block of code
    return value
- Call the function
function_name(p1,p2....)

### Different Ways of declaring User-defined function

```python
# WAP to perform simple calculater +,-,*,/ on 2 numbers
def calculation(a,b):
    inp = input("Choose 1. Addition, 2. Subtraction, 3. Multiplication, 4. Division")
    if inp == '1':
        print(f"Addition: {a} + {b} = {a+b}")
    elif inp == '2':
        print(f"Subtraction: {a} - {b} = {a-b}")
    elif inp == '3':
        print(f"Multiplication: {a} * {b} = {a*b}")
    elif inp == '4':
        print(f"Division: {a} / {b} = {a/b}")
    else:
        print("Invalid input")
```

```python
calculation(10,5)
```

```python
# create all the 4 types of functions for printing prime numbers from 2 to 20, all 4 functions
```

```python
def prime1():
    for i in range(2,21):
        for j in range(2,i):
            if i%j == 0:
                break
        else:
            print(i, end=' ')
```

```python
prime1()
```

```python
def prime2(x1,x2):
    for i in range(x1,x2+1):
        for j in range(x1,i):
            if i%j == 0:
                break
        else:
            print(i, end=' ')
```

```python
prime2(2,20)
```

```python
def prime3():
    primes = []
    for i in range(2,21):
        for j in range(2,i):
            if i%j == 0:
                break
        else:
            primes.append(i)
    return primes
```

```python
prime3()
```

```python
def prime4(x1,x2):
    primes = []
    for i in range(x1,x2+1):
        for j in range(x1,i):
            if i%j == 0:
                break
        else:
            primes.append(i)
    return primes
```

```python
prime4(2,20)
```

```python
# WAF to print fibonacci series numbers without input and without return

def fibonacci():
    num = [0,1]
    while len(num) < 30:
        n = num[-1] + num[-2]
        num.append(n)
    print(num)
```

```python
fibonacci()
```

```python
def fibonacci1():
    n = 30
    a, b = 0, 1
    for i in range(n):
        print(a, end=" ")
        a, b = b, a + b
```

```python
fibonacci1()
```

```python
# recursive function
# A recursive function is a func that calls itself until a condition is met
'''
important parts -
1. base condn - stops recursion
2. recursive call - func calling itself
'''
```

```python
def sum_r(n):
    if n == 1:
        return n
    else:
        return sum_r(n-1)+n
```

```python
sum_r(10)
```

```python
#2 recursive func for factorial of given num
def fact_r(n):
    if n == 1:
        return n
    else:
        return fact_r(n-1)*n
```

```python
fact_r(5)
```
