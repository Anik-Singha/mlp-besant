# Numpy

### Why Numpy ?
It is used for performing operations like array addition, substraction,
multiplication, to create some random numbers, create dummy values etc.

### List vs Array
1. In list, we cannot do element - wise operation but in array we can.
2. Array to List conversion is possible and also List to array conversion is also possible,
but we will lose the dimensions.
3. List is heterogenous datatype but Numpy is a homogenous datatype.
