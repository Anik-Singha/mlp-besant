# File Handling
- It will allow us to create, read, write, and delete files.
- It is used to store data parmenently.

- The built-in function is open() to work with files
- Common nodes:
  1. "r" - read
  2. "w" - write
  3. "a" - append
  4. "b" - binary
  5. "t" - text
  6. "rb" - read binary
  7. "wt" - write text

## File Path
- realtive path : file name
- absolute path : directories(complete loc) + file name
