## Comments and Keywords

### Comments
- Comments are used to **explain the code** and improve **readability**.
- They are **ignored by the Python interpreter** and are not executed.

### Single-Line Comments
- Single-line comments start with the symbol `#`.
- They are used for **short explanations or notes** in the code.

**Example:**
`# This is a comment`

---

### Multi-Line Comments
- Multi-line comments are written using triple quotes `''' '''` or `""" """`.
- They are used for **writing comments across multiple lines or paragraphs**.

**Example:**
`'''
Hi
Hello
'''`

### Keywords

- **Keywords** are reserved words in Python that have special meanings.
- They **cannot be used as variable names, function names, or identifiers**.
- Keywords define the **syntax and structure of the Python language**.

**Examples:** `if`, `for`, `True`, `False`, `import`, `while`, etc.
