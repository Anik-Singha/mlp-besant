```python
'''
1️⃣ Sales Data Analysis

A company recorded daily sales for 30 days.

Create a NumPy array of 30 random sales values between 1000 and 5000 and calculate:

Total sales

Average sales

Top 5 highest sales days
'''
import pandas as pd
import numpy as np

arr_sales = np.random.randint(1000,5001,size=30)
arr_sales
```

```python
# total sales
np.sum(arr_sales)
```

```python
# avg sales
np.average(arr_sales).round(2)
```

```python
np.mean(arr_sales).round(2)
```

```python
# top 5 sales dates
top5 = np.argsort(arr_sales)[-5:][::-1]
top5
```

```python
top5_sales = arr_sales[top5]
top5_sales
```

```python
# top 5 days
print('Top 5 highest sales days')
for i in range(5):
    print(f'Day {top5[i] + 1} = Sales {top5_sales[i]}') # since day strt frm 0 in python indexes
```

```python
'''
2. Outlier Detection
Generate 100 random numbers between 10 and 100.
Find outliers using Z-score method.
'''
arr_rand_nums = np.random.randint(10,101,size=100)
arr_rand_nums
```

```python
# mean and standard deviation
mean = np.mean(arr_rand_nums)
std = np.std(arr_rand_nums)
```

```python
# z-score
z_score = (arr_rand_nums-mean)/std
```

```python
# identify outliers / |Z| > 3 is common threshold
outliers = arr_rand_nums[np.abs(z_score) > 3]
```

```python
print("Data:\n", arr_rand_nums)
print("\nMean:", mean)
print("Standard Deviation:", std)
print("\nZ-scores:\n", z_score)

print("\nOutliers (Z > 3 or Z < -3):", outliers)
```

```python
'''
3. Data Normalization
Given the dataset:
[120, 150, 200, 250, 300]
Normalize the values using Min-Max Scaling (0–1).
'''
```

```python
# mix-max scaling
d = np.array([120, 150, 200, 250, 300])

d_min = np.min(d)
d_max = np.max(d)

# minmax
minmax = (d - d_min)/ (d_max - d_min)
minmax
```

```python
'''
4.  Customer Age Analysis
Customer ages dataset:
[18,22,25,30,35,40,45,50,55,60]
Find:
Mean age
Median age
Standard deviation
'''

age = np.array([18,22,25,30,35,40,45,50,55,60])
mean_age = np.mean(age)
median_age = np.median(age)
std_age = np.std(age)
print(f'Mean age: {mean_age} Median age: {median_age} Standard deviation: {std_age}')
```

```python
'''
5.Matrix Row Filtering
Generate a 10×5 matrix.
Select rows where the row sum is greater than 200.
'''
m_array = np.random.randint(10,100,size = (10,5))
for i in m_array:
    sum_row = np.sum(i)
    if sum_row > 200:
        print(i)
```

```python
'''
6. Feature Scaling
Create an array of 50 random numbers (1–100).
Standardize the data using:
(x - mean) / standard deviation
'''

data = np.random.randint(1,101,size = 50)
m = np.mean(data)
s = np.std(data)

# standardized mean= 0 and sd = 1
standarzied = (data - m)/s
standarzied
```

```python
# new mean
m_new = np.mean(standarzied)
print(m_new)
s_new = np.std(standarzied)
print(s_new)
```
