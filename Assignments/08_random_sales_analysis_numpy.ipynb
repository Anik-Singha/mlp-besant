```python
'''
1️⃣ Sales Data Analysis

A company recorded daily sales for 30 days.

Create a NumPy array of 30 random sales values between 1000 and 5000 and calculate:

Total sales

Average sales

Top 5 highest sales days
'''
import pandas as pd
import numpy as np

arr_sales = np.random.randint(1000,5001,size=30)
arr_sales
```

```python
# avg sales
np.average(arr_sales).round(2)
```

```python
np.mean(arr_sales).round(2)
```

```python
# top 5 sales dates
top5 = np.argsort(arr_sales)[-5:][::-1]
top5
```

```python
top5_sales = arr_sales[top5]
top5_sales
```

```python
# top 5 days
print('Top 5 highest sales days')
for i in range(5):
    print(f'Day {top5[i] + 1} = Sales {top5_sales[i]}') # since day strt frm 0 in python indexes
```

```python
'''
2. Outlier Detection
Generate 100 random numbers between 10 and 100.
Find outliers using Z-score method.
'''
arr_rand_nums = np.random.randint(10,101,size=100)
arr_rand_nums
```

```python
# mean and standard deviation
mean = np.mean(arr_rand_nums)
std = np.std(arr_rand_nums)
```

```python
# z-score
z_score = (arr_rand_nums-mean)/std
```

```python
# identify outliers / |Z| > 3 is common threshold
outliers = arr_rand_nums[np.abs(z_score) > 3]
```

```python
print("Data:\n", arr_rand_nums)
print("\nMean:", mean)
print("Standard Deviation:", std)
print("\nZ-scores:\n", z_score)

print("\nOutliers (Z > 3 or Z < -3):", outliers)
```

```python
'''
3. Data Normalization
Given the dataset:
[120, 150, 200, 250, 300]
Normalize the values using Min-Max Scaling (0–1).
'''
```

```python
# mix-max scaling
d = np.array([120, 150, 200, 250, 300])

d_min = np.min(d)
d_max = np.max(d)

# minmax
minmax = (d - d_min)/ (d_max - d_min)
minmax
```

```python
'''
14. Distance Calculation
Given two vectors:
A = [2,3,4]
B = [5,6,7]
Calculate the Euclidean distance.
'''

# It measures the straight-line distance between two points
import numpy as np
A = np.array([2, 3, 4])
B = np.array([5, 6, 7])
distance = np.linalg.norm(A - B)
print(distance)
```

```python
'''
18. Correlation Calculation
Generate two arrays of 50 random numbers.
Calculate the correlation coefficient between them.
'''
# It tells how strongly two variables move together
# Step 1: Generate data
x = np.random.rand(50)
y = np.random.rand(50)

# Step 2: Correlation
corr_matrix = np.corrcoef(x, y)

print("Correlation Matrix:\n", corr_matrix)

# Extract correlation value
correlation = corr_matrix[0, 1]
print("\nCorrelation Coefficient:", correlation)
```
