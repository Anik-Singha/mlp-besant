```python
# Q1. Calculate Total Sales Amount for each transaction using a new column
total_sales = df_sales['Quantity'] * df_sales['UnitPrice'] - (df_sales['Quantity'] * df_sales['UnitPrice'])*df_sales['Discount']/100
df_sales['total_sales'] = total_sales
df_sales
```

```python
# Q2. Find total sales by each Region.
df_sales.groupby('Region')['total_sales'].sum()
```

```python
# Q3. Display all transactions where Discount is greater than 5%.
df_sales[df_sales['Discount'] > 5]
```

```python
# 4. Group data by Category and find the average UnitPrice for each.
df_sales.groupby('Category')['UnitPrice'].mean()
```

```python
# 5. Find the top 2 SalesPersons with the highest total sales amount.
df_sales.sort_values('total_sales', ascending=False).iloc[:2]
```

```python
# 6. Filter records for only Electronics category where sales are above ₹50,000.
df_sales[(df_sales['Category'] == 'Electronics') & (df_sales['total_sales'] > 50000)]
```

```python
# 7. Sort the dataset by Date and SalesPerson in ascending order.
df_sales.sort_values(['Date','SalesPerson'])
```

```python
# 8. Find the number of transactions for each CustomerType (Regular vs New)
df_sales.groupby('CustomerType').size()
```

```python
df_sales.groupby("CustomerType")["TransactionID"].nunique()
```

```python
# 9. Get the most frequently sold product in each region.
df_sales.groupby('Region')['Product'].apply(lambda x: x.mode().iat[0])
```

```python
df_sales.groupby('Region')['Product'].sum()
```

```python
# 10. Pivot Table: Create a pivot table showing total sales by Region and Category.
pd.pivot_table(df_sales,columns=['Region','Category'] ,  values='total_sales', aggfunc='sum')
```

```python
pd.pivot_table(
    df_sales,
    index="Region",
    columns="Category",
    values="total_sales",
    aggfunc="sum",
    fill_value=0
)
```
