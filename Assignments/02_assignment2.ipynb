```python
# WAP to read student name, id, email, 3 sub marks out of 100 and display the details in below format
"""
ID:
Name:
Email:
Sub1:
Sub2:
Sub3:
Total:
Percentage:
"""

fields = ["ID", "Name", "Email", "Sub1", "Sub2", "Sub3"]
values = {}
for field in fields:
    values[field] = input(f"Enter {field}: ")

for value in values:
    print(f"{value}: {values[value]}")
total = sum(int(values[key]) for key in values if 'Sub' in key)
print(f"Total: {total}")
print(f"Percentage: {round((total / 300) * 100,2)}%")
```
