```python
even_list = []
odd_list = []
prime_list = []
for i in range(1,21):
  if i%2==0:
    even_list.append(i)
  else:
    odd_list.append(i)
  if i==1:
    continue
  for j in range(2,i):
    if i%j==0:
      break
  else:
    prime_list.append(i)
print(even_list)
print(odd_list)
print(prime_list)
```

```python
# String
'''
a string is considered both a buil-in data type in python and a fundamental data structure
represented by  ' '  or str()

- immutable in nature
- iterable
'''
```

```python
s1 = 'python@123?best:64'

# count no. of char, num of digits , sp char
char_count = 0
digit_count = 0
sp_char_count = 0

for i in s1:
    if (ord(i) >= 65 and ord(i) <= 90) or (ord(i) >= 97 and ord(i) <= 122):
        char_count += 1
    elif ord(i) >= 48 and ord(i) <= 57:
        digit_count += 1
    else:
        sp_char_count += 1

print(char_count, digit_count, sp_char_count)
```
