###  **Loops**
- loops is used to repeat a block of code again and again until a condition is met.

#### Types of Loops
1. For loop
    - for loop with sequence
    - for loop with range
2. while loop
3. nested loop
    - for loop inside for loop
    - for loop inside while loop
    - while loop inside while loop
    - while loop insdie for loop


```python
# for loop with a sequence
'''
for itt_var in sequence:
    statements
'''
name = 'Anik'
for i in name:
    print(i)
```

```python
names = ['Anik', 'Jan', 'Januari']
for name in names:
    print(name,', Happy Wishes')
```

```python
s1 = 'python'
for i in range(len(s1)):
    print(f"postion is {i} and character is {s1[i]}")
```

```python
for i in enumerate(s1):
    print(i)
```

```python
'''
for loop with range syntax

for itt_var in range(start,stop,step):
    statements
'''
for i in range(1,11):
    print(i, end = ' ')
```

```python
# even numbers between 1 and 10
for i in range(2,11,2):
    print(i, end = ' ')
```

```python
for i in range(1,11):
    if i % 2 == 0 :
        print(f'{i} = even')
    else:
        print(f'{i} = odd')
```

```python
# WAP to print table of given number
num = 2
for i in range(1,11):
    print(f'{num} x {i}= {i * num}')
```

```python
# sum of first 10 numbers

sum_of_num = 0
for i in range(1,11):
    sum_of_num += i
print(sum_of_num)
```

```python
# Find factorial of given num   e.g: 5! = 5*4*3*2*1
num = 5
fact_num = 1
for i in range(1,num+1):
    fact_num *= i
print(fact_num)
```

## While Loop


```python
'''
initialization
while(cond):
    statements
    inc/dec
'''
```

```python
# print num from 1 to 10
num = 1
while num <= 10:
    print(num)
    num += 1
```

```python
name = 'Anik'
# out = position = 0, char = 'A'
i = 0
while i < len(name):
    print(f"position is {i} and character is {name[i]}")
    i +=1
```

```python
# WAP to find sum of digits of given num
# ex = 1342 1+3+4+2
# write without built in func
value = 1342
out = 0
for i in str(value):
    out += int(i)
print(out)
```

```python
i = 0
out = 0
while i < len(str(value)):
    out += int(str(value)[i])
    i+=1
print(out)
```

```python
value = 1342
out = 0
while value != 0:
    out += value % 10
    value //= 10
print(out)
```

```python
# Chech given num is palindrome or not
# 121 = 121 is palindrome, 123 = 321 is not a plaindrome

num = 121
temp = num
n = ''
while num != 0:
    n += str(num % 10)
    num //= 10
if int(n) == temp:
    print('Palindrome')
else:
    print('Not Palindrome')
```

```python
num = 12221
temp = num
rev = 0

while temp != 0:
    temp1 = temp % 10
    rev = (rev * 10) + temp1
    temp //= 10
if rev == num:
    print('Palindrome')
else:
    print('Not Palindrome')
```
