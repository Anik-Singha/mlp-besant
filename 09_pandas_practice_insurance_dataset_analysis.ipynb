```python
# checking datatypes
df.dtypes
```

```python
# check for null values of each column
df.isna().sum()
```

```python
# duplicates
df.duplicated().sum()
```

```python
# each region wise total expenses
# gender wise avg bmi and expenses
# each region, each children and somker class wise total expenses
df.groupby('region')['expenses'].sum().sort_values(ascending=False)
```

```python
# to get ans in dataframe
df.groupby("region", as_index=False)["expenses"].sum()
```

```python
# gender wise avg bmi and expenses
df.groupby('sex')[['bmi','expenses']].mean().round(2)
```

```python
# each region, each children and somker class wise total expenses
df.groupby(['region','children','smoker'])[['expenses']].sum()
```

```python
# each region, each children and somker class wise total expenses
pd.pivot_table(df,columns=['region','children','smoker'],values='expenses', aggfunc='sum')
```

```python
pd.pivot_table(
    df,
    index=["region", "children"],
    columns="smoker",
    values="expenses",
    aggfunc="sum"
)
```
