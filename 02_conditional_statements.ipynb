## Conditional Statements

- Conditional statements in python are used to make decisions in program.
- They allow the program to execute different blocks of code depending on whether a condition is True or False.

OR
- They help the program to choose what to do next based on a condition

#### Types
1. if
2. if-else
3. if-elif-else
4. nested if else


```python
# if syntax
'''
if(cond):
    print(cond)
'''
```

```python
# eligibility criteria for voting in India
age = int(input("Enter your age: "))
if age >= 18:
    print("You can vote")
```

```python
# if-else
age = int(input("Enter your age: "))
if age >= 18:
    print("You can vote")
else:
    print("You can not vote")
```

```python
# WAP to check the given number is even or odd

num = 22
if num % 2 == 0:
    print("Even")
else:
    print("Odd")
```

```python
# if-elif-else / if-else ladder
'''
if cond:
    print(cond)
elif cond:
    print(cond)
else:
    print(cond)
'''

# or
'''
if cond:
    print(cond)
else:
    if cond:
        print(cond)
    else:
        print(cond)
'''
```

```python
# if-elif-else practice
num = 80
if num >= 90:
    print('Distinction')
elif num >=60:
    print('First Class')
elif num >= 45:
    print('Second Class')
else:
    print('Fail')
```
