```python
# Exception Handling
# we use try-except

'''
Syntax:

try:
    block of code
except:
    block of code
'''
```

```python
try:
    a = 10
    b = '1'
    print(a/b)
except:
    print("Faaah")
```

```python
try:
    num = int(input("Enter a number: "))
    print(10/num)
except ZeroDivisionError:
    print("Division by zero")
except ValueError:
    print("Value error")
```

```python
try:
    num = int(input("Enter a number: "))
    print(10/num)
except Exception as e:
    print(e)
```

```python
try:
    x = int("abc")
except ValueError:
    print("Value error")
except Exception:
    print("General error")
```

```python
try:
    num = 1
    print(10/num)
except ZeroDivisionError:
    print("Division by zero")
except ValueError:
    print("Value error")
else:
    print("No error")
finally:
    print("Finally statement")
```
