```python
from importlib.metadata import distribution
from time import time

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.get_dataset_names()
```

```python
df = sns.load_dataset('tips')
df
```

```python
plt.figure()
plt.hist(df["total_bill"])
plt.title("histogram")
plt.xlabel("Total Bill")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure()
sns.kdeplot(df["total_bill"], fill = True)
plt.title("KDE Plot")
plt.xlabel("Total Bill")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure()
sns.boxplot(x=df['total_bill'])
plt.title('box plot')
plt.xlabel('Total Bill')
plt.show()
```

# Violin Plot:
- combines box plot + distribution
- shows data shape clearly

```python
plt.figure()
sns.violinplot(x = df['total_bill'])
plt.title('violin plot')
plt.xlabel('Total Bill')
plt.show()
```

# Bar plot
- compares values across categories
- easy to understand differences

```python
plt.figure()
sns.barplot(x = 'day', y = 'total_bill', data = df)
plt.title('bar plot')
plt.show()
```

# Pie Chart
- shows proportion/ percentage
- good for quick insights
#### note: use only for small num of categories

```python
plt.figure()
df['day'].value_counts().plot.pie()
plt.title('pie')
plt.show()
```

```python
counts = df['day'].value_counts()

plt.pie(counts, labels=counts.index, autopct='%1.1f%%')
plt.title('pie')
plt.show()
```

# Line plot:
- shows trends over time
- helps to identify increases, decrease, seasonaltiy

ex - sales growth month-wise

```python
plt.plot(df['total_bill'])
plt.title('line plot')
plt.show()
```

# Count Plot
- shows frequency of categories
- useful for categorical analysis

ex- number of customers per city

```python
sns.countplot(x = df['day'])
plt.title('countplot')
plt.show()
```

```python
df['day'].value_counts()
```

# Scatter Plot
- shows relationship between two nnumerical variables
- helps detect corelation and patterns

ex - Advertising Vs Sales

```python
sns.scatterplot(x = df['total_bill'], y = df['tip'],)
plt.title('scatter plot')
plt.show()
```

```python
sns.scatterplot(data=df, x='total_bill', y='tip', hue='day')
plt.title('scatter plot')
plt.show()
```

# Bubbe Sort
- add third variable(size)
- shows more info in one chart

ex - sales vs profit with market size

```python
plt.scatter(df['total_bill'], df['tip'],s = df['size']*30)
plt.title('Bubble plot')
plt.show()
```

# Pair plot
- shows relationship between all variables
- useful in EDA(exploratory Data Analysis)

ex - feature relationships in ML

```python
sns.pairplot(df)
plt.show()
```

# Heatmap
- shows correlation between variables
- helps in feature selection

ex - identify highly correlated features

```python
corr = df.corr(numeric_only=True)
corr.round(2)
```

```python
sns.heatmap(corr,annot=True)
plt.show()
```
